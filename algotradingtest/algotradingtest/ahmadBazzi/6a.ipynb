```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start=datetime.datetime(2020,1,1)
end=datetime.datetime(2020,3,1)

longFig=(17,5)
rectFig=(15,7)
```

```python
tesla=web.DataReader('TSLA','yahoo',start,end)
ford =web.DataReader('F','yahoo',start,end)
gm   =web.DataReader('GM','yahoo',start,end)
google   =web.DataReader('GOOG','yahoo',start,end)
```

```python
google.head()
```

```python
google['Open'].plot(label='GOOG Open',figsize=rectFig)
google['Close'].plot(label='GOOG Close')
google['High'].plot(label='GOOG High')
google['Low'].plot(label='GOOG Low')
plt.legend()
plt.title('Google stock pwices :3c')
plt.ylabel('USD')
plt.show()
```
