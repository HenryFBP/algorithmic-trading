```python
%%markdown

# daily percentage change:

return at time t = (price{t}/price{t-1})-1
    
# $ return_t = \frac{price_t}{price_{t-1}} - 1 $
```

```python
%%markdown

now let's find out what stock is the most "stable" based on this metric
```

```python
def makeDailyReturn(df):
    df['DailyReturn'] = (df['Close'] / df['Close'].shift(1)) - 1

makeDailyReturn(tesla)
makeDailyReturn(ford)
makeDailyReturn(gm)

tesla.head()
```

```python
ford['DailyReturn'].hist(bins=50)
# the fatter this distribution is, the more "volatile" the return is day-to-day.
```

```python
%%markdown

# $Variance[X] = stdev^2_x$
```

```python
tesla['DailyReturn'].hist(bins=100, label='Tesla',alpha=0.5,figsize=[13,6])
gm['DailyReturn'].hist(bins=100, label='GM',alpha=0.5)
ford['DailyReturn'].hist(bins=100, label='Ford',alpha=0.5)
plt.legend()
```

```python
%%markdown

from this, we can see that tesla stocks are more volatile than GM and ford.

The DailyReturn tends to deviate more from about $120 (x axis). 

we can also smooth this histogram out more by using the KDE (kernel density estimate). see below
```

```python
tesla['DailyReturn'].plot(kind='kde', label='Tesla',alpha=0.5,figsize=[13,6])
```
