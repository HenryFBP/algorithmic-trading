```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start=datetime.datetime(2012,1,1)
end=datetime.datetime(2017,1,1)

longFig=(17,5)
rectFig=(15,7)
month=30
```

```python
tesla=web.DataReader('TSLA','yahoo',start,end)
ford =web.DataReader('F','yahoo',start,end)
gm   =web.DataReader('GM','yahoo',start,end)
```

```python
tesla.to_csv("tesla_stock.csv")
ford.to_csv("ford_stock.csv")
gm.to_csv("general_motors_stock.csv")
```

```python
tesla['Open'].plot(label='Tesla',figsize=rectFig)
gm['Open'].plot(label='GM')
ford['Open'].plot(label='Ford')
plt.legend()
plt.ylabel("Stock Price in USD")
plt.title("Stock prices of Tesla, GM, and Ford")
```

```python
tesla['Volume'].plot(label='Tesla',figsize=rectFig)
gm['Volume'].plot(label='GM')
ford['Volume'].plot(label='Ford')
plt.legend()
```
