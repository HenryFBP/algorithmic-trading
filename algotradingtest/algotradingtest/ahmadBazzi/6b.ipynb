````python
%%markdown

# daily percentage change:

return at time t = (price{t}/price{t-1})-1
    
# $ return_t = \frac{price_t}{price_{t-1}} - 1 $
````

````python
%%markdown

now let's find out what stock is the most "stable" based on this metric
````

````python
def makeDailyReturn(df):
    df['DailyReturn'] = (df['Close'] / df['Close'].shift(1)) - 1

makeDailyReturn(tesla)
makeDailyReturn(ford)
makeDailyReturn(gm)

tesla.head()
````

````python
ford['DailyReturn'].hist(bins=50)
# the fatter this distribution is, the more "volatile" the return is day-to-day.
````

````python
%%markdown

# $Variance[X] = stdev^2_x$
````

````python
tesla['DailyReturn'].hist(bins=100, label='Tesla',alpha=0.5,figsize=[13,6])
gm['DailyReturn'].hist(bins=100, label='GM',alpha=0.5,)
ford['DailyReturn'].hist(bins=100, label='Ford',alpha=0.5,)
plt.legend()
````

````python
%%markdown

from this, we can see that tesla stocks are (probably) more volatile than GM and ford.

The DailyReturn tends to deviate more from about $120 (x axis). 

we can also smooth this histogram out more by using the KDE (kernel density estimate). see below
````

````python
tesla['DailyReturn'].plot(kind='kde', label='Tesla',figsize=[13,6])
ford ['DailyReturn'].plot(kind='kde', label='ford')
gm  ['DailyReturn'].plot(kind='kde', label='gm')
plt.legend()
````

````python
%%markdown

note... the reason we see a different scale in the histogram versus in the kernel density estimate, is that KDE is normalized, 
while the histogram is not normalized.

so, the areas under the KDE graphs are constant...in the histogram chart, they are not.
````

````python
%%markdown

insights... ford and GM daily returns are correlated somewhat.

now...on to:

# cumulative return

# $ i_i = (1 + r_t)i_{t-1} $

this is a recursive formula...

You multiple the cumulative return of (day-1) with $ r_t $... 

$ r_t $ is defined above...it is  $ return_t = \frac{price_t}{price_{t-1}} - 1 $

you're basically accumulating the daily percentage change repeatedly, each day.

note...this formula does not take into account stocks that return a dividend.

this is useful compared to daily return, because daily return only shows returns on a day-to-day basis. 

cumReturn shows returns over a span of time.

example:
    
```
Date     Price
1        10
2        15
3        20
4        25

Date     DailyRet        %DailyRet
1        -               -
2        15/10 = 3/2     50%
3        20/15 = 4/3     33%
4        25/20 = 5/4     20%

Date     CumRet          %CumRet
1        -               -
2        15/10 = 3/2     150%
3        20/10 = 4/2     200%
4        25/10 = 5/2     250%


```
````

````python
tesla['Cumulative Returns'] = (1 + tesla['DailyReturn']).cumprod()

tesla['Cumulative Returns'].head()
````

````python
tesla['DailyReturn'].head()
````

````python
ford ['Cumulative Returns'] = (1 + ford ['DailyReturn']).cumprod()
gm   ['Cumulative Returns'] = (1 + gm   ['DailyReturn']).cumprod()

ford.head()

# note the first value is always close to 1... this is because we start out with a return of "1" at day 0,
# because we just started tracking what our investment would make.
````

````python
tesla['Cumulative Returns'].plot(label='tesla',figsize=(15,6))
ford ['Cumulative Returns'].plot(label='ford' ,figsize=(15,6))
gm   ['Cumulative Returns'].plot(label='gm'   ,figsize=(15,6))
plt.title('Cumulative Return vs time')
plt.legend()
````
